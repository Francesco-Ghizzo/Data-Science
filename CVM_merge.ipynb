Imports

```python
import pandas as pd
import numpy as np
```

Constants

```python
CD_CONTA = 'Código da Conta'
DS_CONTA = 'Descrição da Conta'
```

User defined functions

```python
def get_len_key(df):
    return [len(key) for key in df[CD_CONTA]]
```

```python
def get_min_len(df):
    len_key = get_len_key(df)
    return np.array(len_key).min()
```

```python
def get_ind_min_len(df):
    len_key = get_len_key(df)
    min_len = get_min_len(df)
    return np.where(len_key==min_len)
```

```python
def get_ind_min_len(df, min_len):
    len_key = get_len_key(df)
    return np.where(len_key==min_len)
```
