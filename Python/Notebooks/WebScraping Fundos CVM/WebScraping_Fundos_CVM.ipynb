```python
import requests
import lxml
from bs4 import BeautifulSoup
import html5lib
import pandas as pd
```

```python
url = "https://cvmweb.cvm.gov.br/SWB/Sistemas/SCW/CPublica/CDA/CPublicaCDA.aspx?PK_PARTIC=165919&SemFrame="                                   # get page from a given url
response = requests.get(url) 
```

```python
encoding = response.encoding                                    
page = response.content.decode(encoding)                      # decode page content in non human readable form
```

```python
formatted_page = BeautifulSoup(response.text, 'lxml')                     # format page in a human readable form
```

```python
table = pd.read_html(page, attrs = {'id': 'dlAplics'})                    # read table from html using pandas. Select table using id
table = table[0]
table
```
