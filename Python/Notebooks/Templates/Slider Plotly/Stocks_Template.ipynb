Imports

````python
import pandas as pd
import plotly.graph_objects as go
````

```
Load datasets
```

````python
try:
    AAPL = pd.read_csv('AAPL_data.csv')
    print("Data series loaded!")
except:
    print("Error: Failed to load data series!")
````

````python
try:
    GOOGL = pd.read_csv('GOOGL_data.csv')
    print("Data series loaded!")
except:
    print("Error: Failed to load data series!")
````

```
Read Column Names
```

````python
AAPL.head()
````

```
Interactive Plot
```

````python
df = AAPL
x_col = 'date'
y_col = 'close'
plot_title = 'AAPL'
````

````python
fig = go.Figure([go.Scatter(x=df[x_col], y=df[y_col])])
fig.update_layout(title_text=plot_title)
fig.show()
````

```
Interactive plot with slider
```

````python
fig = go.Figure([go.Scatter(x=df[x_col], y=df[y_col])])
fig.update_layout(title_text=plot_title, xaxis_rangeslider_visible=True)
fig.show()
````

````python
fig = go.Figure()
fig.add_trace(go.Scatter(x=df[x_col], y=df[y_col]))
fig.update_layout(title_text=plot_title, xaxis_rangeslider_visible=True)
fig.show()
````

```
Two Data Series Interactive Plot
```

````python
df1 = AAPL
df2 = GOOGL
ts1_name = 'AAPL'
ts2_name = 'GOOGL'
x_col = 'date'
y_col = 'close'
````

````python
fig = go.Figure()
fig.add_trace(go.Scatter(x=df1[x_col], y=df1[y_col], name=ts1_name))
fig.add_trace(go.Scatter(x=df2[x_col], y=df2[y_col], name=ts2_name))
fig.show()
````

```
Two Data Series Interactive Plot with Slider
```

````python
fig = go.Figure()
fig.add_trace(go.Scatter(x=df1[x_col], y=df1[y_col], name=ts1_name))
fig.add_trace(go.Scatter(x=df2[x_col], y=df2[y_col], name=ts2_name))
fig.update_layout(xaxis_rangeslider_visible=True)
fig.show()
````
