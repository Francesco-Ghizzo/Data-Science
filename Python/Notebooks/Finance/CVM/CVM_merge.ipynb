Imports

```python
import pandas as pd
import numpy as np
```

Constants

```python
CD_CONTA = 'Código da Conta'
```

User defined functions

```python
def higher_tier_rows(df):
    len_key = [len(key) for key in df[CD_CONTA]]   # len_key = length of key
    min_len = np.array(len_key).min()  # min_len = minimum length
    ind_min_len = np.where(len_key==min_len)  # ind_min_len = index of key 'Código da Conta' with minimum length
    return ind_min_len
```
