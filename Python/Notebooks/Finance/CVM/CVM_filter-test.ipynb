```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'dfp_cia_aberta_BPA_con_2020.csv', encoding = 'ISO-8859-1', sep=";")    # leio o .csv
```

```python
df = df[df['CNPJ_CIA'] == '97.837.181/0001-47']    # dataframe = linhas do dataframe onde ('CNPJ_CIA' == '97.837.181/0001-47') == True
                                                # (seleciono somente as linhas relativas a uma comp de interesse)
```

```python
df
```

```python
df = df[['ORDEM_EXERC','CD_CONTA','DS_CONTA','VL_CONTA']].copy()
```

```python
df
```

```python
df.drop_duplicates(['CD_CONTA','DS_CONTA'], inplace = True)
```

```python
df
```

```python
CD_DS_CONTA_dict = {a:b for a,b in zip(df['CD_CONTA'],df['DS_CONTA'])}    # crio um dicionário chamado "CD_DS_CONTA_dict" com os elementos de 'CD_CONTA' como keys e os elementos de 'DS_CONTA'como values
```

```python
CD_DS_CONTA_dict
```

```python
def filter(code, len_min_code):
    higher_tier = CD_DS_CONTA_dict[code[:len_min_code]]
    return higher_tier
```

```python
len_min_code = 1
df.insert(loc=2, column='1º Classificação', value=df['CD_CONTA'].map(lambda x: filter(x, len_min_code)))   # df.insert(loc, column, value)
```
