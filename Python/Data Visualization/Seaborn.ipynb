```python
import pandas as pd
import seaborn as sbn
```

```python
flights = sbn.load_dataset('flights')
```

```python
flights
```

```python
sbn.relplot(data=flights, x='year', y='passengers', hue='month', kind='line')
```

```python
flights_wide = flights.pivot(index='year', columns='month', values='passengers')
flights_wide
```
