```python
long_df = pd.read_csv('ch_03/data/long_data.csv', parse_dates=['date'], usecols=['datatype', 'date', 'value'])
```

```python
long_df = long_df[['date', 'datatype', 'value']]
long_df
```

```python
sbn.set(rc={'figure.figsize':(15, 5)}, style='white')
sbn.lineplot(data=long_df, x='date', y='value', hue='datatype')
```

```python
wide_df = long_df.pivot(index='date', columns='datatype', values='value')
wide_df
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
ax.plot_date(wide_df.index.to_pydatetime(), wide_df['TOBS'], '-')
ax.fill_between(wide_df.index.to_pydatetime(), wide_df['TMIN'], wide_df['TMAX'], alpha=0.2)
plt.show()
```
