<h1>How do I select a subset of a <code>DataFrame?</code></h1>

<h2>How do I select specific columns from a <code><span>DataFrame</span></code>?</h2>

image.png

<p>I’m interested in the age of the Titanic passengers.</p>

```python
ages = titanic["Age"]
```

```python
ages.head()
```

<p>To select a single column, use square brackets <code>[]</code> with the column
name of the column of interest.</p>

<p>Each column in a <a href="../../reference/api/pandas.DataFrame.html#pandas.DataFrame" title="pandas.DataFrame"><code>DataFrame</code></a> is a <a href="../../reference/api/pandas.Series.html#pandas.Series" title="pandas.Series"><code>Series</code></a>.
As a single column is selected, the returned object is a pandas <a href="../../reference/api/pandas.Series.html#pandas.Series" title="pandas.Series"><code>Series</code></a>.
We can verify this by checking the type of the output:</p>

```python
type(titanic["Age"])
```

<p>And have a look at the <code>shape</code> of the output:</p>

```python
titanic['Age'].shape
```

<p><a href="../../reference/api/pandas.DataFrame.shape.html#pandas.DataFrame.shape" title="pandas.DataFrame.shape"><code>DataFrame.shape</code></a> is an attribute (remember <a href="02_read_write.html#min-tut-02-read-write">tutorial on reading and writing</a>, do not use parentheses for attributes) of a pandas <code>Series</code> and <code>DataFrame</code> containing the number of rows and
columns: <em>(nrows, ncolumns)</em>.
A pandas Series is 1-dimensional and only the number of rows is returned.</p>

<p>I’m interested in the age and sex of the Titanic passengers.</p>

```python
age_sex = titanic[["Age", "Sex"]]
```

```python
age_sex.head()
```

<p>To select multiple columns, use a list of column names within the
selection brackets <code>[]</code>.</p>

<p>Note</p>
<p>The inner square brackets define a <a href="https://docs.python.org/3/tutorial/datastructures.html#tut-morelists" title="(in Python v3.13)">Python list</a> with column names, whereas the outer brackets are used to select the data from a pandas <code>DataFrame</code> as seen in the previous example.</p>

<p>The returned data type is a pandas DataFrame:</p>

```python
type(titanic[["Age", "Sex"]])
```

```python
titanic[["Age", "Sex"]].shape
```

<p>The selection returned a <code>DataFrame</code> with 891 rows and 2 columns.
Remember, a <code>DataFrame</code> is 2-dimensional with both a row and column dimension.</p>

<h2>How do I filter specific rows from a <code>DataFrame</code>?</h2>

image.png

<p>I’m interested in the passengers older than 35 years.</p>

```python
above_35 = titanic[titanic["Age"] > 35]
```

```python
above_35.head()
```

<p>To select rows based on a conditional expression, use a condition inside
the selection brackets <code>[]</code>.</p>

<p>The condition inside the selection brackets <code>titanic["Age"] &gt; 35</code> checks for which rows the <code>Age</code>column has a value larger than 35:</p>

```python
titanic["Age"] > 35
```

<p>The output of the conditional expression (<code>&gt;</code>, but also <code>==</code>, <code>!=</code>, <code>&lt;</code>, <code>&lt;=</code>,… would work) is actually a pandas <code>Series</code> of boolean values (either <code>True</code> or <code>False</code>) with the same number of rows as the original <code>DataFrame</code>.
Such a <code>Series</code> of boolean values can be used to filter the <code>DataFrame</code> by putting it in between the selection brackets <code>[]</code>.
Only rows for which the value is <code>True</code> will be selected.</p>

<p>We know from before that the original Titanic <code>DataFrame</code> consists of 891 rows.
Let’s have a look at the number of rows which satisfy the
condition by checking the <code>shape</code> attribute of the resulting
<code>DataFrame</code> <code>above_35</code>:</p>

```python
above_35.shape
```

<p>I’m interested in the Titanic passengers from cabin class 2 and 3.</p>

```python
class_23 = titanic[titanic["Pclass"].isin([2, 3])]
```

```python
class_23.head()
```

<p>Note</p>

<p>When combining multiple conditional statements, each condition
must be surrounded by parentheses <code>()</code>.
Moreover, you can not use <code>or</code>/<code>and</code> but need to use the <code>or</code> operator <code>|</code> and the <code>and</code>
operator <code>&amp;</code>.</p>

To user guide

<p>See the dedicated section in the user guide about <a href="../../user_guide/indexing.html#indexing-boolean">boolean indexing</a> or about the <a href="../../user_guide/indexing.html#indexing-basics-indexing-isin">isin function</a>.</p>

<p>I want to work with passenger data for which the age is known.</p>

<p>The <a href="../../reference/api/pandas.Series.notna.html#pandas.Series.notna" title="pandas.Series.notna"><code>notna()</code></a> conditional function returns a <code>True</code> for each row the values are not a <code>Null</code> value.
As such, this can be combined with the selection brackets <code>[]</code> to filter the data table.</p>

```python
age_no_na = titanic[titanic['Age'].notna()]
```

```python
age_no_na.head()
```

<p>The <a href="../../reference/api/pandas.Series.notna.html#pandas.Series.notna" title="pandas.Series.notna"><code>notna()</code></a> conditional function returns a <code>True</code> for each row the values are not a <code>Null</code> value.
As such, this can be combined with the selection brackets <code>[]</code> to filter the data table.</p>

<p>You might wonder what actually changed, as the first 5 lines are still
the same values. One way to verify is to check if the shape has changed:</p>

```python
age_no_na.shape
```
